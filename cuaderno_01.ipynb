```r
library(rvest)
```

```r
html <- read_html("https://www.lanacion.com.ar/")
```

```r
print(html)
```

```r
body <- html_element(html, "body")
```

```r
print(body)
```

```r
urls <- html_elements(body, "a") |> html_attr("href")
```

```r
print(urls)
```

```r
titulos <- html_elements(body, "h1, h2, h3, h4, h5, h6")
```

```r
print(titulos)
```

```r
titulos_txt <- titulos |> html_text2()
```

```r
titulos_txt |> print()
```

```r
parrafos <- html_elements(body, "p")
```

```r
print(parrafos)
```

```r
parrafos_txt <- parrafos |> html_text2()
```

```r
for(p in parrafos_txt) { cat(p, "\n") }
```
