```r
library(rvest)
```

```r
html <- read_html("https://www.masrionegro.com/secciones/sindicales/page/1/")
```

```r
body <- html_element(html, "body")
```

```r
titulos <- html_elements(body, "h1, h2, h3, h4, h5, h6") |> html_text2()
```

```r
print(titulos)
```

```r
parrafos <- html_elements(body, "p") |> html_text2()
```

```r
parrafos
```

```r
stringr::str_trim(parrafos)
```
