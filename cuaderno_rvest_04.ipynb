```r
library(tidyverse)
library(rvest)
url <- "https://datos.gob.ar/dataset/habitat-registro-nacional-barrios-populares"
html <- read_html(url)
links <- url_absolute(html_attr(html_elements(html, "a"), "href"), url)
walk(links, ~cat(.x, "\n"))
```

```r
links_csv <- links[str_detect(links, "csv")]
walk(links_csv, ~cat(.x, "\n"))
```

```r
barrios_populares <- read_csv(links_csv[1])
```
